```python
import torch
from transformers import pipeline

DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

pipe = pipeline("text-generation", model="jphme/Llama-2-13b-chat-german", max_new_tokens=100)
```

```python
question = "Was ist Ihre Lieblingsfarbe?"
prompt = "<s>[INST] <<SYS>> You are a helpful, respectful and honest assistant. Always answer as helpfully as possible, while being safe. Your answers should not include any harmful, unethical, racist, sexist, toxic, dangerous, or illegal content. Please ensure that your responses are socially unbiased and positive in nature. If a question does not make any sense, or is not factually coherent, explain why instead of answering something not correct. If you don't know the answer to a question, please don't share false information. Please answer in the same language as the user. <</SYS>> " + question + "[/INST] "
print("USER: " + question)
print("BOT:  " + pipe(prompt)[len(prompt):])
```

```python
question = "Was denkst du über Brasilien?"
prompt = "<s>[INST] <<SYS>> You are a friendly German language partner. Your role is to chat with language learners to help them improve their language skills. Your answers should be brief, friendly and conversational. Please answer in B1-level German. <</SYS>> " + question + "[/INST]"
print("USER: " + question)
print("BOT: " + pipe(prompt)[0]['generated_text'][len(prompt):])
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("jphme/Llama-2-13b-chat-german")
model = AutoModelForCausalLM.from_pretrained("jphme/Llama-2-13b-chat-german")
```

```python
import json

# Load the constraint list JSON object
with open('constraint_lists.json', 'r', encoding="utf-8") as f:
    constraint_lists = json.load(f)

# Make list from A1 Unit 1
test_list = constraint_lists["A1 Full List"]["Unit 1"]

# Tada!
print(test_list)
```

```python
force_words = test_list[:30]
force_words_ids = [
    tokenizer(force_words, add_special_tokens=False).input_ids,
]

for word in force_words:
    print(word)
```

```python
question = "Hallo!"
prompt = "<s>[INST] <<SYS>> You are a friendly German language partner. Your role is to chat with language learners to help them improve their language skills. Your answers should be brief, friendly and conversational. Please answer in A1-level German. <</SYS>> " + question + "[/INST]"
print("USER: " + question)

input_ids = tokenizer(prompt, return_tensors="pt").input_ids
outputs = model.generate(
    input_ids,
    force_words_ids=force_words_ids,
    num_beams=10,
    num_return_sequences=1,
    no_repeat_ngram_size=1,
    remove_invalid_values=True,
    max_new_tokens=100,
)

print("BOT: " + tokenizer.decode(outputs[0])[len(prompt):])
```
